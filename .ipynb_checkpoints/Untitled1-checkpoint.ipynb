```python
import numpy as np
import time
import meshcat
from roboticstoolbox import DHRobot, RevoluteDH , ctraj
from spatialmath import SE3
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


#checking if our notebook is readding our env kernal 
print("Everything is working!")
```

```python
vis = meshcat.Visualizer().open() 
```

```python
# define Dh parameter for DOBOT CR10A
j1a=0
j1alpha=np.pi/2
j1d=0.1765

j2a=0.607
j2alpha=0
j2d=0

j3a=0.568
j3alpha=0
j3d=0

j4a=0
j4alpha=np.pi/2
j4d=0.125

j5a=0
j5alpha=-np.pi/2
j5d=0.125

j6a=0
j6alpha=0
j6d=0.1114
```

```python
robot = DHRobot([
    RevoluteDH(d=j1d, a=j1a, alpha=j1alpha),
    RevoluteDH(d=j2d, a=j2a, alpha=j2alpha),
    RevoluteDH(d=j3d, a=j3a, alpha=j3alpha),
    RevoluteDH(d=j4d, a=j4a, alpha=j4alpha),
    RevoluteDH(d=j5d, a=j5a, alpha=j5alpha),
    RevoluteDH(d=j6d, a=j6a, alpha=j6alpha)
], name="Dobot CR10A")
```

```python
q_start = [0, 0, 0, 0, 0, 0]
q_end   = [0.3, -0.3, 0.2, 0.1, 0.2, -0.1]
```

```python
T0 = robot.fkine(q_start)
T1 = robot.fkine(q_end)
```

```python
traj = ctraj(T0, T1, 50)
```

```python
robot.plot(q_start, backend='meshcat', viewer=vis)
```
