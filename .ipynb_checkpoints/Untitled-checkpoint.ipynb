```python
import numpy as np
import pandas as pd

import time

from roboticstoolbox import DHRobot, RevoluteDH , ctraj , jtraj
from spatialmath import SE3
from sympy import Matrix , pprint
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


#checking if our notebook is readding our env kernal 
print("Everything is working!")
```

## defining the robot 
 in the arrangement j1 -----  joint one 
 and also total mass of the robot is 40 since we are focusing on the joints and 55-60% of the mass is mass of the base
  and mass of actuator we distrbute 40% of the mass for the joints 

```python
# define Dh parameter for DOBOT CR10A
j1a=0
j1alpha=np.pi/2
j1d=0.1765
m1=3.5

j2a=0.607
j2alpha=0
j2d=0
m2=3.0

j3a=0.568
j3alpha=0
j3d=0
m3=2.8

j4a=0
j4alpha=np.pi/2
j4d=0.125
m4=2.5

j5a=0
j5alpha=-np.pi/2
j5d=0.125
m5=2.0

j6a=0
j6alpha=0
j6d=0.1114
m6=2.2
```

## constracting the robot 
 we use the dhrobot () class imported from roboticstoolbox to constract the robot object defined above  

```python
robot = DHRobot([
    RevoluteDH(d=j1d, a=j1a, alpha=j1alpha , m=m1),
    RevoluteDH(d=j2d, a=j2a, alpha=j2alpha , m=m2),
    RevoluteDH(d=j3d, a=j3a, alpha=j3alpha , m=m3),
    RevoluteDH(d=j4d, a=j4a, alpha=j4alpha , m=m4),
    RevoluteDH(d=j5d, a=j5a, alpha=j5alpha , m=m5),
    RevoluteDH(d=j6d, a=j6a, alpha=j6alpha , m=m6)
], name="Dobot CR10A")
```

## forward kinematics 
we use test angel q to demonstrate the forward kinematics of our robot 
the result is calculated using the fkine methode defined under dhrobot class 

```python
q = [np.pi/2, 0, -np.pi/6, 0, 0, 0]   #  test angels 
T = robot.fkine(q)
print("Forward Kinematics of DOBOT CR10A end effectors postion  :")
print(T)
```

## inverse kinematics 

we use desired postion t to calculate our angel q 
we use the ikin_LM() methode defined under dhrobot class

```python
T_desired = SE3(0.5, 0.0, 0.2)  # test postion 


ik = robot.ikine_LM(T_desired)
print("Inverse Kinematics of DOBOT CR10A joint angel in rad ")
print(ik.q)
```

## jackobina and dynamics of the robot 

we deined postion m velocity v and accleration a to calculate respective jackobia matrix and dynamics of the robot 

```python
m=[0, -np.pi/6, np.pi/6, 0, 0, 0]
v=[0.1 , -0.1 , 0.3 , 0.4 , 0.1 , -0.1]
a=[0.1 , 0.4 , 0.3 , -0.1 , 0.1 , 0.466]
```

```python

jacobian=robot.jacob0(m)
```

```python

print("jacobian matrix " , jacobian)
```

```python
# Linear velocity part (top 3 rows)
Jv = pd.DataFrame(jacobian.iloc[:3, :])

# Angular velocity part (bottom 3 rows)
Jw = pd.DataFrame(jacobian.iloc[3:, :])


print("Linear velocity part (Jv)")
Jv


```

```python
print("Angular velocity part (Jw)")
Jw
```

```python
pd.DataFrame(jacobian)
```

```python
fig, ax = plt.subplots()
cax = ax.matshow(jacobian)
fig.colorbar(cax)

for (i, j), val in np.ndenumerate(jacobian):
    ax.text(j, i, f"{val:.2f}", ha='center', va='center')

ax.set_title("6x6 Jacobian Matrix")
ax.set_xlabel("Joint Index")
ax.set_ylabel("Task Space (vx,vy,vz,wx,wy,wz)")
plt.show()
```

## trajectory 

```python
q_start=[-np.pi/2, 0, 0, np.pi/6, 0, np.pi/4]
q_end=[0, 0, 0, 0, 0, -np.pi/2]
steps=100
t=0.4
```

```python
traj=jtraj(q_start , q_end , steps)
```

```python
mt=traj.q
```

```python
vt=traj.qdd
```

```python
mt
```

```python
time=np.linspace(0,t,steps)
```

```python
plt.figure(figsize=(10,6))
for j in range(mt.shape[1]):
    plt.plot(time , mt[:, j] , label=f'joint {j+1}')

plt.title(" dobot cr10a")
plt.xlabel("time")
plt.ylabel("postion")
plt.legend()
plt.grid(True)
plt.show()
```
